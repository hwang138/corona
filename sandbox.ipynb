### Visualization of the latest Cornona-19 data


This report refreshes daily and provides latest case and death numbers from<br>
European Centre for Disease Prevention and Control [(ecdc)](
https://www.ecdc.europa.eu/en/geographical-distribution-2019-ncov-cases
)

To donwload the csv file: [link](
https://www.ecdc.europa.eu/en/publications-data/download-todays-data-geographic-distribution-covid-19-cases-worldwide
)

Currently only reporting on these countries:
- United_States_of_America
- Taiwan
- Italy
- Spain
- Germany
- China


```python
from corona.corona import Corona

%load_ext nb_black
%load_ext autoreload
%autoreload 2
```

```python
corona = Corona()
```

```python
initial_y_list = [
    "cases",
    "deaths",
]
per_cap_y_list = [elem + "_per_capita" for elem in initial_y_list]
cumulative_y_list = ["cumulative_" + elem for elem in initial_y_list]
cumulative_per_cap_y_list = [elem + "_per_capita" for elem in cumulative_y_list]

for y_list in [
    cumulative_y_list,
    cumulative_per_cap_y_list,
    initial_y_list,
    per_cap_y_list,
]:
    subset_country_list = [
        "United_States_of_America",
        "Taiwan",
        "Italy",
        "Spain",
        "Germany",
        "China",
    ]
    corona.plotting_utils.plot_time_series(
        data=corona.data.query("countries_and_territories in @subset_country_list"),
        x="date",
        y=y_list,
        ax_factor="countries_and_territories",
        fontsize=12,
    )
    print("\n\n")
```
